```python
# ✅ 1. 留下需要的欄位
df = df[['sitename', 'county', 'pollutant', 'status', 'aqi']]
print("留下欄位後的資料：")
display(df.head())

# ✅ 2. 找出空氣品質不良的測站（AQI > 100）
bad_air_df = df[df['aqi'] > 100]
print("AQI > 100 的測站：")
display(bad_air_df)

# ✅ 3. 查詢污染指標物有哪些（pollutant 不重複值）
pollutants = df['pollutant'].unique()
print("污染指標物有：")
print(pollutants)

# ✅ 4. 找出高雄市的測站資料
kaohsiung_df = df[df['county'] == '高雄市']
print("高雄市的測站資料：")
display(kaohsiung_df)

# ✅ 5. 計算各縣市的平均 AQI
county_avg_aqi = df.groupby('county')['aqi'].mean().sort_values(ascending=False)
print("各縣市平均 AQI：")
display(county_avg_aqi)
```
